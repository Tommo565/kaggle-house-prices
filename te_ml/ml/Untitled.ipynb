```python
import pandas as pd
```

```python
def import_files(train_in, test_in):
    '''Imports the train and test sets'''

    df_train = pd.read_csv(train_in)
    df_train['Train'] = 1

    df_test = pd.read_csv(test_in)
    df_test['Test'] = 1

    df_all = pd.concat([df_train, df_test])

    return df_train, df_test, df_all
```

```python
df_train, df_test, df_all = import_files('../data/input_data/train.csv', '../data/input_data/test.csv')
```
