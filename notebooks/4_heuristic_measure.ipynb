```python
import numpy as np
import pandas as pd
from math import sqrt
```

```python
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', -1)
```

```python
infile = './transformed_data/train_heuristic.csv'
```

```python
df = pd.read_csv(infile)
df.head()
```

```python
# Taking the mean of the best predictor and using that as the predictor

tab = df[['OverallQual', 'SalePrice']].groupby(['OverallQual']).mean().reset_index()
tab.rename(columns={'SalePrice':'SalePricePred'}, inplace=True)
```

```python
df = pd.merge(
    left=df,
    right=tab,
    left_on='OverallQual',
    right_on='OverallQual',
    how='left'
)
```

```python
def RMSE(row):
    diff_sq = sqrt((row['SalePrice'] - row['SalePricePred'])**2)
    return diff_sq

df['RMSE'] = df.apply(RMSE, axis=1)
df['RMSE_log'] = np.log(df['RMSE'])
result = df['RMSE_log'].mean()
result
```
